```python
#Importação das bibliotecas utilizadas na lista
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import warnings
from scipy.fft import fft, fftshift
import math
from sympy import Symbol, symbols
warnings.filterwarnings('ignore')
```

```python
def impD(n): #função impulso
  y = np.double(n == 0)
  return y

def uD(n):
  y = np.double(n >= 0)
  return y
```

```python
#Plotando um círculo unitário no plano Z
def circulo(titulo,r):
  figure, axes = plt.subplots()

  draw_circle = plt.Circle((0, 0), r,fill=False)

  axes.set_aspect(1)
  axes.add_artist(draw_circle)
  plt.title(titulo)
  plt.xlabel('Im{z}')
  plt.ylabel('Re{z}')

  axes.add_artist(lines.Line2D([0,0],[-5*r,5*r],color="#000000"))
  axes.add_artist(lines.Line2D([5*r,-5*r],[0,0],color="#000000"))

  plt.plot(1,0,color="#000000")
  plt.plot(-1,0,color="#000000")
  plt.plot(0,1,color="#000000")
  plt.plot(0,-1,color="#000000")
```

```python
# Plotagem da região e de outros componentes
def regiao(raio,operador,cor_f = 'red',fig='',ax=''):
  figure, axes = plt.subplots()

  if fig != '' and ax!="":
    figure = fig
    axes = ax

  cor_fora = 'white'
  if operador == '<':
    prenc = '///'
  elif operador == '>':
    prenc = None
    cor_fora = cor_f

  draw_circle = plt.Circle((0, 0),raio ,linestyle='--',linewidth=1,edgecolor=cor_f,hatch=prenc,facecolor='white',zorder=10,in_layout=False)

  lim = raio+2*raio


  plt.fill_between([-lim,lim],lim , hatch='/', edgecolor=cor_fora,facecolor='white')
  plt.fill_between([-lim,lim],-lim, hatch="/", edgecolor=cor_fora,facecolor='white')

  axes.add_artist(lines.Line2D([0,0],[-lim,lim],color="#000000",zorder=11))
  axes.add_artist(lines.Line2D([lim,-lim],[0,0],color="#000000",zorder=11))


  plt.xlim(-lim,lim)
  plt.ylim(-lim,lim)

  plt.scatter(0,0,s=2*raio,c='red',zorder=11)

  axes.set_aspect(1)
  axes.add_artist(draw_circle)
  axes.set_xlabel('Im{z}')
  axes.set_ylabel('Re{z}')


  return figure,axes
```

### 2

image.png

<font color = 'blue'>Para essa transformada Z inversa, iremos aplicar a tecnica de frações parciais para facilitar a observação da função.

<font color = 'blue'>Primeiro temos que:

$$z^2-2z+1 = (z-1)(z-1);$$

<font color = 'blue'>Aplicando essa condição em X(z) temos:

$$X(z) = \frac{z-1}{(z-1)(z-1)};$$

<font color = 'blue'>Fazendo as devidas fatorações temos que:

$$X(z) = \frac{1}{(z-1);}$$

<font color = 'blue'>Verificando a tabela dos pares de transformada, não encontramos nenhuma transformada Z que satisfaça a expressão encontrada, logo, iremos manipulá-la para chegar a um valor que nos possibilita aplicar a tabela:

<font color = 'blue'>Multiplicando o numerador e o denominador por $z^{-1}$ temos:

$$X(z) = \frac{z^{-1}}{(z-1)z^{-1}};$$

$$X(z) = z^{-1}\frac{1}{(1-z^{-1})};$$

$$X(z) = z^{-1}\frac{z}{(z-1)};$$

<font color = 'blue'>O resultado observado é caracterizado pelo produto de duas transformadas, nesse caso, fora do domínio Z, temos a convolução dos seguintes sinais:

$$δ[n-m] <TZ> z^{-1};$$
e 
$$u[n] <TZ> \frac{z}{z-1};$$

<font color = 'blue'>por fim temos que a transformada Z inversa de X(z) é:

$$Xz^{-1}\frac{z}{(z-1)}>TZ>u[n]*δ[n-1].$$


```python
n  = np.arange(-10,11,1) # definindo um intervalo qualquer

u1=np.zeros_like(n)
u1[n>=0]=1 # declarando a função impulso
sigma = impD(n-1) # declarando a função impulso

xn = np.convolve(u1,sigma,'same') # fazendo a convolução dos dois sinais

plt.stem(n,xn) # plot da convolução dos sinais

plt.title("Figura 8 - Sinal da transformada Z inversa de X(z) ")
plt.xlabel("n")
plt.ylabel("x[n]")

plt.show()
```

<font color = 'blue'>Sabemos que a transformada Z inversa de  $z^{-m}$ é $δ[n-m]$, contudo, nada podemos afirmar da outra expressão, mas sabemos que:

```python
n  = np.arange(-10,11,1) # definindo um intervalo qualquer
u1=np.zeros_like(n)
u1[n>=0]=1 # declarando a função impulso
x1 = ((1/2)**n)*u1 # declarando a função impulso
sigma2 = impD(n-2) # declarando a função degrau
xn2 = np.convolve(x1,sigma2,'same') # fazendo a convolução dos dois sinais
plt.stem(n,xn2)

plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Figura 9 - Sinal da de x[n] ')

plt.show()
```

<font color = 'blue'>Para que um sistema H(z) seja estável, ele deve obedecer ao critério tal que tendo uma função de tranferência:

image.png

<font color = 'blue'>No qual $p_k$ são os polos de H(z), temos que o critério é:

$$|p_k|< 1, k= 0,1,2,3,4,...$$

<font color = 'blue'>Esse critério é baseado no comportamento de que sistemas em tempo discretos são estáveis se todos os seus pólos estiverem no interior do circulo unitário do plano-Z. Isso pois polos que estão dentro do círculo unitário contibuem para um decaimento exponencial da resposta ao impulso, o que é muito importante, visto que para um sistema ser estável o sinal precisa ser convergente, logo, não pode ter termos exponenciais absolutamente crescentes. Para questão de entendimento, após obter a resposta, vou conferir graficamente o resultado.

<font color = 'blue'>Dessa forma, vamos fazer uma manipulação em H(z) para que fique mais claro os polos da função.

$$H(z)= \frac{K}{1+K·\frac{z}{z-0.9}}$$

<font color = 'blue'>Obtendo a seguinte expressão:

$$H(z)= \frac{\frac{K}{K+1}·(z-0.9)}{z-\frac{0.9}{K+1}}$$

<font color = 'blue'>Dessa forma, fica fácil observar que o sistema tem apenas um polo:

$$p_k = \frac{0.9}{K+1}$$

<font color = 'blue'>E para que ele seja menor que 1, basta que:

$$K>-0.1$$



```python
#Plotando um círculo unitário no plano Z
titulo = 'Figura 19 - Representação do cículo unitário e dos polos da função H(z) no plano Z, para diferentes valores de K.'
circulo(titulo,1)

#Plot dos pólos no plano Z
K = Symbol('K')

Ks = [-0.3,-0.2,-0.1,0,1] # Valores de K que eu escolhi para observar o comportamento dos polos no plano Z
names = ['K=-0.3','K=-0.2','K=-0.1','K=0','K=1'] # Legenda
colors = ['red','blue','orange','purple','cadetblue'] # Cor para ficar bonito

for i in range(5):
  K = Ks[i] # Atribuição dos valores escolhidos para K
  polo = 0.9/(K+1) #Polo da função H(z)
  plt.plot(polo,0,marker='x',color=colors[i],label=names[i],markersize=12)

plt.legend()
plt.show()
```

<font color = 'blue'>Como se vê na figura acima, quando K=-0.1 o polo fica na borda do círculo unitário, a partir do momento em que K>-0.1, os polos já estão dentro do círculo unitário caracterizando um sistema estável, já para valores de K<-0.1, é possível observar que temos um polo fora do círculo, logo o sistema se torna instável.
