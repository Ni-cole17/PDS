```python
#Importao das bibliotecas utilizadas na lista
import numpy as np
from scipy import *
from scipy import signal
import matplotlib.pyplot as plt
import warnings
from scipy.fft import fft, fftshift
import math
from scipy import integrate
warnings.filterwarnings('ignore')
```

```python
def uC(t):
  y = np.double(t >= 0);
  return y

def retC(T,t):
  y = uC(t + T/2) - uC(t - T/2);
  return y

def rmpC(t): #funo rampa
  y = t*np.double(t >= 0)
  return y

def impTC(T,t):
  y = np.double(t/T == np.round(t/T))
  return y;
```

**Para aplicar o teorema de Parseval,  necessrio que seja utilizada a transformada de x[n] que  a sua forma em: $X(e^{jw})$. Primeiramente, para encontrar a transformada decidi dividir x[n] em duas partes (I) e (II) e calcular a transformada de cada uma separadamente. A primeira  formada pela funo sinc e a segunda  formada pela funo seno, como mostra abaixo.** 

$(I)  $  $ sinc(\frac{n}{10}) = 10ret(\frac{10w}{2})*_{2}(w) $

$(II) $  $ sen(\frac{n}{2}) = j[_{2}(w+\frac{}{2})-_{2}(w-\frac{}{2})]$

**Depois de achada as suas transformadas, utilizou-se a seguinte propriedade para que achar a transformada de x[n]:**
$$x[n] = x_{(I)}[n]x_{(II)}[n] = \frac{1}{2}X_{(I)}(e^{jw})*X_{(II)}(e^{jw})$$
**Logo, temos que:**
$$X(e^{jw}) = \frac{1}{2}[10ret(\frac{5w}{})*_{2}(w)*j[_{2}(w+\frac{}{2})-_{2}(w-\frac{}{2})]] $$

**Analisando essa equao, utilizando ainda as equaes de $x_I$ e $x_{II}$  possvel observar que:**

**Por meio da equao $(I)$  esperado que seja um sinal retangular de largura $\frac{}{5}$, com amplitude de 10, e centrado em $w_0=0$, visto que o trem de impulso est centrado em $w_0=0$, como mostrado na figura abaixo.**

```python
w = np.arange(-20*np.pi/20,21*np.pi/20,np.pi/20) # Perodo de integrao de -pi at pi (esse perodo vai ser utilizado em todo o desenvolvimento do exerccio)

ret = retC(np.pi/5,w) # Funo retangular com largura pi/5
imp = impTC(2*np.pi,w) # Funo impulso com perodo de 2pi
sinc = 10*np.convolve(ret,imp,'same') # Convoluo entre a funo retangular e a funo impulso

#Plot do sinal da transformada de x[n] = sinc(n/10)
plt.plot(w,sinc)
plt.title('Sinal da transformada de fourier de x[n] = sinc(n/10)')
plt.ylabel('x(e^jw)')
plt.xlabel('w')

plt.show()
```

**Vimos que obtivemos o resultado esperado para a transformada da primeira equao $(I)$. Analisando agora a segunda equao $(II)$, temos dois trens de impulso centrados em $\frac{-}{2}$ e em $\frac{}{2}$, multiplicados por uma constante complexa. Como o objetivo  ver o *plot*, vou considerar o mesmo intervalo e o valor absoluto da constante.**

```python
#Valor absoluto das funes impulso, com objetivo de fazer o plot 
imp1 = abs(np.complex(0,1)*np.pi)*impTC(2*np.pi,w+(np.pi/2)) 
imp2 = abs(np.complex(0,1)*np.pi)*impTC(2*np.pi,w-(np.pi/2))
sen = imp1 - imp2 #Transformada de sen(2*pi*n/4)

#Plot da transformada
plt.plot(w,sen)
plt.title('Sinal da transformada de fourier de x[n] = sen(pi*n/2)')
plt.ylabel('x(e^jw)')
plt.xlabel('w')

plt.show()
```

**Portanto, o resultado esperado para $X(e^{jw})$ que  a convoluo entre $x_I$ e $x_{II}$, que ter como resultado dois pulsos retangulares de largura $\frac{pi}{5}$ centrados em $\frac{-pi}{2}$ e $\frac{pi}{2}$** 

```python
def x(w): #Funo para calcular a transformada de fourier de x[n], dado um determinado w.

  #Transformada da funo sinc
  ret = retC(np.pi/5,w)
  imp = impTC(2*np.pi,w)
  sinc = 10*np.convolve(ret,imp,'same')

  #Transformada da funo sen
  imp1 = impTC(2*np.pi,w+(np.pi/2))
  imp2 = impTC(2*np.pi,w-(np.pi/2))
  sen = imp1 - imp2

  #Transformada de x[n] (convoluo entre as transformadas acima)
  sinal = abs((1/(2*np.complex(0,1))))*np.convolve(sen,sinc,'same') #Pegando o valor absoluto da constante para representar no grfico

  return sinal
```

**Pode-se observar que o resultado do sinal da transformada na figura abaixo foi o resultado esperado**

**Funo que calcula a DFT**

```python
def calcularDFT (w0, fs, N):
  # Definindo o perdo de amostragem
  ts = 1/fs

  # Vetor do domnio
  t = np.arange(0, N) * ts


  # Sinal a ser calculada a DFT
  x = np.cos(w0*t)

  # Vetor no domnio K
  df = 1/N*ts
  k = np.arange (-N/2, N/2) * df

  # Clculo da transformada
  xf = fftshift(fft(x))

  # Plot e ajustes grficos
  figure, (ax1, ax2) = plt.subplots (2,1)
  ax1.stem (k, np.abs(xf))
  ax1.set_title ("Mdulo da Tranformada de Fourier X(f)")

  ax2.stem (k, np.angle(xf))
  ax2.set_title ("Fase da Tranformada de Fourier X(f)")

  plt.subplots_adjust(top=0.952, hspace=0.452)
```

Como os sinais da questo so de tempo contnuo e peridicos, podemos considerar a relao entre o valor absoluto e a sua fase conforme equao:

Capturar.PNG

Nesse caso, temos a representao em termos desse valor absoluto e fase das transformadas de fourier para tempo contnuo. A o mdulo da transformada $X(j)$ descreve o contedo de frequncia bsico de um sinal, ou seja, $X(j)$ proporciona a informao sobre as magnitudes relativas das exponenciais complexas que compem x(t) na sua transformada, representado pelo grfico do mdulo da transformada de Fourier X(f). A fase $X(j)$, por outro lado, no afeta o mdulo dos componentes de frequncia individuais, mas se refere os angulos de fases relativas dessas exponenciais, representado pelo grfico da fase da transformada de Fourier X(f). As relaes de fase representadas por $X(j)$ possuem um efeito significativo sobre a natureza do sinal x(t) e, portanto, tipicamente contm uma quantidade substancial de informaes sobre o sinal.
