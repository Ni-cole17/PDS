```python
#Importação das bibliotecas utilizadas na lista
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import warnings
from scipy.fft import fft, fftshift
import math
warnings.filterwarnings('ignore')
```

```python
#Funções que serão utilizadas ao longo do código
def impND(N,n): #Função do trem de impulso
  if N == round(N):
    y = np.double(n/N==np.round(n/N))
  return y

def soma(x): #função soma de vetor
  y = np.cumsum(x)
  return y

def rmpC(t): #função rampa
  y = t*np.double(t >= 0)
  return y

def retD(N,n): #função porta
  if N == round(N):
    y = np.double(abs(n)<=abs(N))
  return y

def impD(n): #função impulso
  y = np.double(n == 0)
  return y
```

Para calcular o período do gráfico 1, sabemos que como X[n] é uma função periódica, seu período pode ser calculado da seguinte forma:
$$T = \frac{2π}{b}$$
Sendo b toda a parte que multiplica n, dentro da função cosseno.
Portanto como temos: $$cos(\frac{2πn}{10})  →  T = \frac{2π}{\frac{2π}{10}} →  T = 10$$ 
Ao compararmos o resultado calculado com o resultado obtido no gráfico 2, temos que o período é 10 para ambos. 

Para a frequência fundamental, utilizamos a seguinte expressão para o calculo:
$$f_o = \frac{1}{T}$$
Como T = 10, $$f_o = \frac{1}{10} Hz$$

O período da função X[n], nesse caso, será calculada multiplicando os períodos das funções cossseno que fazem parte da soma. Primeiramente, vamos calcular o período da primeira função cosseno:
$$cos(2πn/5) →  T_1 = \frac{2π}{\frac{2π}{5}} →  T_1 = 5 $$
Para a segunda função:
$$cos(2πn/7) →  T_2 = \frac{2π}{\frac{2π}{7}} →  T_2 = 7 $$
Assim, para o período da função X[n]:
$$T_x = T_1 ⋅ T_2 = 35$$
Esse resultado foi o mesmo para o observado na função X[n], como mostrado no gráfico 4.

Para a frequência fundamental
$$f_o = \frac{1}{T_x}$$
Como $T_x = 35$, $$f_o = \frac{1}{35} Hz$$

O perído da função X[n] é a soma de duas exponenciais com argumento complexo, logo podemos utilizar a identidade de euler para facilitar os cálculos do período.
$$e^{\frac{2nπj}{20}} = cos(\frac{2nπ}{20}) + j⋅sen(\frac{2nπj}{20})$$
$$e^{\frac{-2nπj}{20}} = cos(\frac{2nπ}{20}) - j⋅sen(\frac{2nπj}{20})$$
Somando as duas exponenciais temos que:
$$X[n] = 2⋅cos(\frac{2nπ}{20})$$
Logo, chegamos em uma função trigonométrica periódica (cossseno), sendo assim podemos aplicar a fórmula do período $T = \frac{2π}{b}$.

Assim, temos que o período da função X[n] é:
$$T_x = 20$$
Observando o período do gráfico 6, percebe-se que o resultado calculado e observado foi o mesmo.

Por fim, para a frequência fundamental, como $T_x = 20$:
 $$f_o = \frac{1}{20} Hz$$
