```python
#Importao das bibliotecas utilizadas na lista
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import warnings
from scipy.fft import fft, fftshift
import math
warnings.filterwarnings('ignore')
```

```python
#Funes que sero utilizadas ao longo do cdigo
def impND(N,n): #Funo do trem de impulso
  if N == round(N):
    y = np.double(n/N==np.round(n/N))
  return y

def soma(x): #funo soma de vetor
  y = np.cumsum(x)
  return y

def rmpC(t): #funo rampa
  y = t*np.double(t >= 0)
  return y

def retD(N,n): #funo porta
  if N == round(N):
    y = np.double(abs(n)<=abs(N))
  return y

def impD(n): #funo impulso
  y = np.double(n == 0)
  return y
```

Para calcular o perodo do grfico 1, sabemos que como X[n]  uma funo peridica, seu perodo pode ser calculado da seguinte forma:
$$T = \frac{2}{b}$$
Sendo b toda a parte que multiplica n, dentro da funo cosseno.
Portanto como temos: $$cos(\frac{2n}{10})    T = \frac{2}{\frac{2}{10}}   T = 10$$ 
Ao compararmos o resultado calculado com o resultado obtido no grfico 2, temos que o perodo  10 para ambos. 

Para a frequncia fundamental, utilizamos a seguinte expresso para o calculo:
$$f_o = \frac{1}{T}$$
Como T = 10, $$f_o = \frac{1}{10} Hz$$

O perodo da funo X[n], nesse caso, ser calculada multiplicando os perodos das funes cossseno que fazem parte da soma. Primeiramente, vamos calcular o perodo da primeira funo cosseno:
$$cos(2n/5)   T_1 = \frac{2}{\frac{2}{5}}   T_1 = 5 $$
Para a segunda funo:
$$cos(2n/7)   T_2 = \frac{2}{\frac{2}{7}}   T_2 = 7 $$
Assim, para o perodo da funo X[n]:
$$T_x = T_1  T_2 = 35$$
Esse resultado foi o mesmo para o observado na funo X[n], como mostrado no grfico 4.

Para a frequncia fundamental
$$f_o = \frac{1}{T_x}$$
Como $T_x = 35$, $$f_o = \frac{1}{35} Hz$$

O perdo da funo X[n]  a soma de duas exponenciais com argumento complexo, logo podemos utilizar a identidade de euler para facilitar os clculos do perodo.
$$e^{\frac{2nj}{20}} = cos(\frac{2n}{20}) + jsen(\frac{2nj}{20})$$
$$e^{\frac{-2nj}{20}} = cos(\frac{2n}{20}) - jsen(\frac{2nj}{20})$$
Somando as duas exponenciais temos que:
$$X[n] = 2cos(\frac{2n}{20})$$
Logo, chegamos em uma funo trigonomtrica peridica (cossseno), sendo assim podemos aplicar a frmula do perodo $T = \frac{2}{b}$.

Assim, temos que o perodo da funo X[n] :
$$T_x = 20$$
Observando o perodo do grfico 6, percebe-se que o resultado calculado e observado foi o mesmo.

Por fim, para a frequncia fundamental, como $T_x = 20$:
 $$f_o = \frac{1}{20} Hz$$
