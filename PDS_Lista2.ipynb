```python
#Importação das bibliotecas utilizadas na lista
import numpy as np
from scipy import *
from scipy import signal
import matplotlib.pyplot as plt
import warnings
from scipy.fft import fft, fftshift
import math
from scipy import integrate
warnings.filterwarnings('ignore')
```

```python
def uC(t):
  y = np.double(t >= 0);
  return y

def retC(T,t):
  y = uC(t + T/2) - uC(t - T/2);
  return y

def rmpC(t): #função rampa
  y = t*np.double(t >= 0)
  return y

def impTC(T,t):
  y = np.double(t/T == np.round(t/T))
  return y;
```

**Sabemos que a definição do somatório para a TFTD é dada da seguinte forma:**

$$X(e^{jw}) = \sum \limits_{i=-\infty}^{\infty}x[n]e^{-jwn}$$

**E sabendo que o sinal x[n] é do tipo porta, tendo a TFTD ,por meio da tabela, o formato:**
$$Y(e^{jw})=(2N+1).sinc((2N+1)\frac{w}{2π})*δ_{2π}(w)$$

**temos, por fim, que a TFTD do sinal x[n] é:**

$$Y(e^{jw})=90.sinc(9\frac{w}{2π})*δ_{2π}(w)$$ 
**ou**
$$Y(e^{jw})=90.drcl(\frac{w}{2π},9)$$ 


**Para esse e para os próximos itens, decidimos representar algumas transformadas de fourier, como uma forma de facilitar o entendimento do resultado por nós.**

```python
#Codigo da transformada item 1.
w = np.arange(-20*np.pi/20,21*np.pi/20,np.pi/20) #intervalo
x = 90*np.convolve(np.sinc(9*w/(2*np.pi)),impTC(2*np.pi,w),'same') #Transformada de Fourier

#Representação gráfica da transformada
plt.plot(w,x)
plt.title("Gráfico 1 - Representação da transformada do sinal x[n]")
plt.xlabel("w")
plt.ylabel("Y(e^jw)")

plt.show()
```

**Para aplicar o teorema de Parseval, é necessário que seja utilizada a transformada de x[n] que é a sua forma em: $X(e^{jw})$. Primeiramente, para encontrar a transformada decidi dividir x[n] em duas partes (I) e (II) e calcular a transformada de cada uma separadamente. A primeira é formada pela função sinc e a segunda é formada pela função seno, como mostra abaixo.** 

$(I)  $  $ sinc(\frac{n}{10}) = 10ret(\frac{10w}{2π})*δ_{2π}(w) $

$(II) $  $ sen(\frac{πn}{2}) = jπ[δ_{2π}(w+\frac{π}{2})-δ_{2π}(w-\frac{π}{2})]$

**Depois de achada as suas transformadas, utilizou-se a seguinte propriedade para que achar a transformada de x[n]:**
$$x[n] = x_{(I)}[n]x_{(II)}[n] = \frac{1}{2π}X_{(I)}(e^{jw})*X_{(II)}(e^{jw})$$
**Logo, temos que:**
$$X(e^{jw}) = \frac{1}{2π}·[10ret(\frac{5w}{π})*δ_{2π}(w)*jπ[δ_{2π}(w+\frac{π}{2})-δ_{2π}(w-\frac{π}{2})]] $$

**Analisando essa equação, utilizando ainda as equações de $x_I$ e $x_{II}$ é possível observar que:**

**Por meio da equação $(I)$ é esperado que seja um sinal retangular de largura $\frac{π}{5}$, com amplitude de 10, e centrado em $w_0=0$, visto que o trem de impulso está centrado em $w_0=0$, como mostrado na figura abaixo.**

```python
w = np.arange(-20*np.pi/20,21*np.pi/20,np.pi/20) # Período de integração de -pi até pi (esse período vai ser utilizado em todo o desenvolvimento do exercício)

ret = retC(np.pi/5,w) # Função retangular com largura pi/5
imp = impTC(2*np.pi,w) # Função impulso com período de 2pi
sinc = 10*np.convolve(ret,imp,'same') # Convolução entre a função retangular e a função impulso

#Plot do sinal da transformada de x[n] = sinc(n/10)
plt.plot(w,sinc)
plt.title('Sinal da transformada de fourier de x[n] = sinc(n/10)')
plt.ylabel('x(e^jw)')
plt.xlabel('w')

plt.show()
```

**Vimos que obtivemos o resultado esperado para a transformada da primeira equação $(I)$. Analisando agora a segunda equação $(II)$, temos dois trens de impulso centrados em $\frac{-π}{2}$ e em $\frac{π}{2}$, multiplicados por uma constante complexa. Como o objetivo é ver o *plot*, vou considerar o mesmo intervalo e o valor absoluto da constante.**

```python
#Valor absoluto das funções impulso, com objetivo de fazer o plot 
imp1 = abs(np.complex(0,1)*np.pi)*impTC(2*np.pi,w+(np.pi/2)) 
imp2 = abs(np.complex(0,1)*np.pi)*impTC(2*np.pi,w-(np.pi/2))
sen = imp1 - imp2 #Transformada de sen(2*pi*n/4)

#Plot da transformada
plt.plot(w,sen)
plt.title('Sinal da transformada de fourier de x[n] = sen(pi*n/2)')
plt.ylabel('x(e^jw)')
plt.xlabel('w')

plt.show()
```

**Portanto, o resultado esperado para $X(e^{jw})$ que é a convolução entre $x_I$ e $x_{II}$, que terá como resultado dois pulsos retangulares de largura $\frac{pi}{5}$ centrados em $\frac{-pi}{2}$ e $\frac{pi}{2}$** 

```python
def x(w): #Função para calcular a transformada de fourier de x[n], dado um determinado w.

  #Transformada da função sinc
  ret = retC(np.pi/5,w)
  imp = impTC(2*np.pi,w)
  sinc = 10*np.convolve(ret,imp,'same')

  #Transformada da função sen
  imp1 = impTC(2*np.pi,w+(np.pi/2))
  imp2 = impTC(2*np.pi,w-(np.pi/2))
  sen = imp1 - imp2

  #Transformada de x[n] (convolução entre as transformadas acima)
  sinal = abs((1/(2*np.complex(0,1))))*np.convolve(sen,sinc,'same') #Pegando o valor absoluto da constante para representar no gráfico

  return sinal
```

**Função que calcula a DFT**

```python
def calcularDFT (w0, fs, N):
  # Definindo o perído de amostragem
  ts = 1/fs

  # Vetor do domínio
  t = np.arange(0, N) * ts


  # Sinal a ser calculada a DFT
  x = np.cos(w0*t)

  # Vetor no domínio K
  df = 1/N*ts
  k = np.arange (-N/2, N/2) * df

  # Cálculo da transformada
  xf = fftshift(fft(x))

  # Plot e ajustes gráficos
  figure, (ax1, ax2) = plt.subplots (2,1)
  ax1.stem (k, np.abs(xf))
  ax1.set_title ("Módulo da Tranformada de Fourier X(f)")

  ax2.stem (k, np.angle(xf))
  ax2.set_title ("Fase da Tranformada de Fourier X(f)")

  plt.subplots_adjust(top=0.952, hspace=0.452)
```

Como os sinais da questão são de tempo contínuo e periódicos, podemos considerar a relação entre o valor absoluto e a sua fase conforme equação:

Capturar.PNG

Nesse caso, temos a representação em termos desse valor absoluto e fase das transformadas de fourier para tempo contínuo. A o módulo da transformada $X(jω)$ descreve o conteúdo de frequência básico de um sinal, ou seja, $X(jω)$ proporciona a informação sobre as magnitudes relativas das exponenciais complexas que compõem x(t) na sua transformada, representado pelo gráfico do módulo da transformada de Fourier X(f). A fase $X(jω)$, por outro lado, não afeta o módulo dos componentes de frequência individuais, mas se refere àos angulos de fases relativas dessas exponenciais, representado pelo gráfico da fase da transformada de Fourier X(f). As relações de fase representadas por $X(jω)$ possuem um efeito significativo sobre a natureza do sinal x(t) e, portanto, tipicamente contêm uma quantidade substancial de informações sobre o sinal.
