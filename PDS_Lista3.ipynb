```python
#Importao das bibliotecas utilizadas na lista
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import warnings
from scipy.fft import fft, fftshift
import math
from sympy import Symbol, symbols
warnings.filterwarnings('ignore')
```

```python
def impD(n): #funo impulso
  y = np.double(n == 0)
  return y

def uD(n):
  y = np.double(n >= 0)
  return y
```

```python
#Plotando um crculo unitrio no plano Z
def circulo(titulo,r):
  figure, axes = plt.subplots()

  draw_circle = plt.Circle((0, 0), r,fill=False)

  axes.set_aspect(1)
  axes.add_artist(draw_circle)
  plt.title(titulo)
  plt.xlabel('Im{z}')
  plt.ylabel('Re{z}')

  axes.add_artist(lines.Line2D([0,0],[-5*r,5*r],color="#000000"))
  axes.add_artist(lines.Line2D([5*r,-5*r],[0,0],color="#000000"))

  plt.plot(1,0,color="#000000")
  plt.plot(-1,0,color="#000000")
  plt.plot(0,1,color="#000000")
  plt.plot(0,-1,color="#000000")
```

```python
# Plotagem da regio e de outros componentes
def regiao(raio,operador,cor_f = 'red',fig='',ax=''):
  figure, axes = plt.subplots()

  if fig != '' and ax!="":
    figure = fig
    axes = ax

  cor_fora = 'white'
  if operador == '<':
    prenc = '///'
  elif operador == '>':
    prenc = None
    cor_fora = cor_f

  draw_circle = plt.Circle((0, 0),raio ,linestyle='--',linewidth=1,edgecolor=cor_f,hatch=prenc,facecolor='white',zorder=10,in_layout=False)

  lim = raio+2*raio


  plt.fill_between([-lim,lim],lim , hatch='/', edgecolor=cor_fora,facecolor='white')
  plt.fill_between([-lim,lim],-lim, hatch="/", edgecolor=cor_fora,facecolor='white')

  axes.add_artist(lines.Line2D([0,0],[-lim,lim],color="#000000",zorder=11))
  axes.add_artist(lines.Line2D([lim,-lim],[0,0],color="#000000",zorder=11))


  plt.xlim(-lim,lim)
  plt.ylim(-lim,lim)

  plt.scatter(0,0,s=2*raio,c='red',zorder=11)

  axes.set_aspect(1)
  axes.add_artist(draw_circle)
  axes.set_xlabel('Im{z}')
  axes.set_ylabel('Re{z}')


  return figure,axes
```

### 2

image.png

<font color = 'blue'>Para essa transformada Z inversa, iremos aplicar a tecnica de fraes parciais para facilitar a observao da funo.

<font color = 'blue'>Primeiro temos que:

$$z^2-2z+1 = (z-1)(z-1);$$

<font color = 'blue'>Aplicando essa condio em X(z) temos:

$$X(z) = \frac{z-1}{(z-1)(z-1)};$$

<font color = 'blue'>Fazendo as devidas fatoraes temos que:

$$X(z) = \frac{1}{(z-1);}$$

<font color = 'blue'>Verificando a tabela dos pares de transformada, no encontramos nenhuma transformada Z que satisfaa a expresso encontrada, logo, iremos manipul-la para chegar a um valor que nos possibilita aplicar a tabela:

<font color = 'blue'>Multiplicando o numerador e o denominador por $z^{-1}$ temos:

$$X(z) = \frac{z^{-1}}{(z-1)z^{-1}};$$

$$X(z) = z^{-1}\frac{1}{(1-z^{-1})};$$

$$X(z) = z^{-1}\frac{z}{(z-1)};$$

<font color = 'blue'>O resultado observado  caracterizado pelo produto de duas transformadas, nesse caso, fora do domnio Z, temos a convoluo dos seguintes sinais:

$$[n-m] <TZ> z^{-1};$$
e 
$$u[n] <TZ> \frac{z}{z-1};$$

<font color = 'blue'>por fim temos que a transformada Z inversa de X(z) :

$$Xz^{-1}\frac{z}{(z-1)}>TZ>u[n]*[n-1].$$


```python
n  = np.arange(-10,11,1) # definindo um intervalo qualquer

u1=np.zeros_like(n)
u1[n>=0]=1 # declarando a funo impulso
sigma = impD(n-1) # declarando a funo impulso

xn = np.convolve(u1,sigma,'same') # fazendo a convoluo dos dois sinais

plt.stem(n,xn) # plot da convoluo dos sinais

plt.title("Figura 8 - Sinal da transformada Z inversa de X(z) ")
plt.xlabel("n")
plt.ylabel("x[n]")

plt.show()
```

<font color = 'blue'>Sabemos que a transformada Z inversa de  $z^{-m}$  $[n-m]$, contudo, nada podemos afirmar da outra expresso, mas sabemos que:

```python
n  = np.arange(-10,11,1) # definindo um intervalo qualquer
u1=np.zeros_like(n)
u1[n>=0]=1 # declarando a funo impulso
x1 = ((1/2)**n)*u1 # declarando a funo impulso
sigma2 = impD(n-2) # declarando a funo degrau
xn2 = np.convolve(x1,sigma2,'same') # fazendo a convoluo dos dois sinais
plt.stem(n,xn2)

plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Figura 9 - Sinal da de x[n] ')

plt.show()
```

<font color = 'blue'>Alm disso, podemos obter os polos do sistema conforme realizado no cdigo acima, a partir da funo [K,p,k]= signal.residue(B,A).

<font color = 'blue'>Por fim, fazendo os clculos da funo de transferncia obtemos os polos

<font color = 'blue'>Para que um sistema H(z) seja estvel, ele deve obedecer ao critrio tal que tendo uma funo de tranferncia:

image.png

<font color = 'blue'>No qual $p_k$ so os polos de H(z), temos que o critrio :

$$|p_k|< 1, k= 0,1,2,3,4,...$$

<font color = 'blue'>Esse critrio  baseado no comportamento de que sistemas em tempo discretos so estveis se todos os seus plos estiverem no interior do circulo unitrio do plano-Z. Isso pois polos que esto dentro do crculo unitrio contibuem para um decaimento exponencial da resposta ao impulso, o que  muito importante, visto que para um sistema ser estvel o sinal precisa ser convergente, logo, no pode ter termos exponenciais absolutamente crescentes. Para questo de entendimento, aps obter a resposta, vou conferir graficamente o resultado.

<font color = 'blue'>Dessa forma, vamos fazer uma manipulao em H(z) para que fique mais claro os polos da funo.

$$H(z)= \frac{K}{1+K\frac{z}{z-0.9}}$$

<font color = 'blue'>Obtendo a seguinte expresso:

$$H(z)= \frac{\frac{K}{K+1}(z-0.9)}{z-\frac{0.9}{K+1}}$$

<font color = 'blue'>Dessa forma, fica fcil observar que o sistema tem apenas um polo:

$$p_k = \frac{0.9}{K+1}$$

<font color = 'blue'>E para que ele seja menor que 1, basta que:

$$K>-0.1$$



```python
#Plotando um crculo unitrio no plano Z
titulo = 'Figura 19 - Representao do cculo unitrio e dos polos da funo H(z) no plano Z, para diferentes valores de K.'
circulo(titulo,1)

#Plot dos plos no plano Z
K = Symbol('K')

Ks = [-0.3,-0.2,-0.1,0,1] # Valores de K que eu escolhi para observar o comportamento dos polos no plano Z
names = ['K=-0.3','K=-0.2','K=-0.1','K=0','K=1'] # Legenda
colors = ['red','blue','orange','purple','cadetblue'] # Cor para ficar bonito

for i in range(5):
  K = Ks[i] # Atribuio dos valores escolhidos para K
  polo = 0.9/(K+1) #Polo da funo H(z)
  plt.plot(polo,0,marker='x',color=colors[i],label=names[i],markersize=12)

plt.legend()
plt.show()
```

<font color = 'blue'>Como se v na figura acima, quando K=-0.1 o polo fica na borda do crculo unitrio, a partir do momento em que K>-0.1, os polos j esto dentro do crculo unitrio caracterizando um sistema estvel, j para valores de K<-0.1,  possvel observar que temos um polo fora do crculo, logo o sistema se torna instvel.
